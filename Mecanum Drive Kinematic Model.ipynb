# Mecanum Drive Kinematic Model
Omnidirectional motion (movement in any direction).Each wheel is mounted at a 45-degree angle 

1. **Assumptions**
 * The robot has four wheels: Front-Left (FL), Front-Right (FR), Rear-Left (RL), and Rear-Right (RR).
 * The robot can move in the x (forward/backward), y (side-to-side), and θ (rotation) directions simultaneously.

Each wheel’s velocity will affect both v_x, v_y, and ω.

v_x : Linear velocity in the x-direction.

v_x : Linear velocity in the y-direction.

ω : Angular velocity.

# Robot velocities based on wheel velocities

a : Half distance between  front and rear wheels (1/2 length).

b : Half distance between the left and right wheels (1/2 width).

r : radius of the wheel.

Assume ω_FL, ω_FR , ω_RL and ω_RR

```python
a = 0.2
b = 0.15
r = 0.03

ω_FL = 3.66
ω_FR = 11
ω_RL = 11
ω_RR = 3.66

v_FL = ω_FL * r
v_FR = ω_FR * r
v_RL = ω_RL * r
v_RR = ω_RR * r

v_x = ( v_FL + v_FR + v_RL + v_RR)/4
v_y = ( -v_FL + v_FR + v_RL - v_RR)/4
ω = ( -v_FL + v_FR - v_RL + v_RR)/(4*(a+b))

print(f"Vx : {v_x}")
print(f"Vy : {v_y}")
print(f"w  : {ω}")
```

# Inverse Kinematics (Calculate Wheel Velocities)


```python
V_FL = v_x - v_y - ω*(a+b)
V_FR = v_x + v_y + ω*(a+b)
V_RL = v_x + v_y - ω*(a+b)
V_RR = v_x - v_y + ω*(a+b)

ω_FL = V_FL / r
ω_FR = V_FR / r
ω_RL = V_RL / r
ω_RR = V_RR / r

print(f"vFL : {V_FL}")
print(f"vFR : {V_FR}")
print(f"vRL  : {V_RL}")
print(f"vRR  : {V_RR}")

print(f"ωFL : {ω_FL}")
print(f"ωFR : {ω_FR}")
print(f"ωRL  : {ω_RL}")
print(f"ωRR  : {ω_RR}")
```
